##### Perform Feature Standerd Scalling
Standardize features by removing the mean and scaling to unit variance

The standard score of a sample x is calculated as:

> z = (x - u) / s

```python
# Scale the data to keep all the values in the same magnitude of 0 -1

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

#### Making the Ml model function

```python
def model_building(model, X_train, X_test, y_train, y_test):
    """
    Model Fitting, Prediction And Other stuff
    return ('score', 'accuracy_score', 'predictions' )
    """

    model.fit(X_train, y_train)
    score = model.score(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(predictions, y_test)

    return (score, accuracy, predictions)
```

```python
models_list = {
    "LogisticRegression": LogisticRegression(),
    "RandomForestClassifier": RandomForestClassifier(
        n_estimators=10, criterion="entropy", random_state=5
    ),
    "DecisionTreeClassifier": DecisionTreeClassifier(
        criterion="entropy", random_state=0
    ),
    "SVC": SVC(),
}
```

```python
# classification
df_prediction = []
confusion_matrixs = []
df_prediction_cols = ["model_name", "score", "accuracy_score", "accuracy_percentage"]

for name, model in zip(list(models_list.keys()), list(models_list.values())):
    (score, accuracy, predictions) = model_building(
        model, X_train, X_test, y_train, y_test
    )

    print("\n\nClassification Report of '" + str(name), "'\n")

    print(classification_report(y_test, predictions))

    df_prediction.append([name, score, accuracy, "{0:.2%}".format(accuracy)])

    # For Showing Metrics
    confusion_matrixs.append(confusion_matrix(y_test, predictions))


df_pred = pd.DataFrame(df_prediction, columns=df_prediction_cols)
```

predict a single output

```python
def model_building_predict_single(model, X_train, y_train):
    """
    Train the model and return a function for making predictions on new input
    """

    model.fit(X_train, y_train)

    def predict_single_input(X_single):
        """
        Predict the output for a single set of input features
        """
        return model.predict([X_single])

    return predict_single_input


# info: In this modified version, model_building returns a function predict_single_input. This function takes a single set of input features X_single and returns the corresponding prediction using the trained model.
```

```python

#  creating the prediction function by calling model_building_predict_single
predict_function = model_building_predict_single(
    models_list.get("RandomForestClassifier"), X_train, y_train
)


first_entry_X_test = X_train[0]

# Using the predict_function to make predictions on the first entry of X_test
prediction = predict_function(first_entry_X_test)


def result(prediction):
    if prediction[0] == 1:
        return "Malignant"
    return "Benign"


print("Prediction:", result(prediction))
```
