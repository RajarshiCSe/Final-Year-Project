```python
import matplotlib.pyplot as plt

# Accuracy values
models = ["Base", "VGG", "ResNet", "DenseNet", "EfficientNet"]
accuracies = [0.87, 0.831, 0.859, 0.833, 0.852]

plt.figure(figsize=(10, 6))
plt.bar(models, accuracies, color="lightgreen")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Comparison of Model Accuracies")
plt.ylim(0.8, 1.0)  # Set y-axis limits
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Accuracy values
models_acc = ["Base", "VGG", "ResNet", "DenseNet", "EfficientNet"]
accuracies = [0.87, 0.831, 0.859, 0.833, 0.852]

# ROC scores
models_roc = ["Base Model", "VGG16", "ResNet50", "DenseNet121", "EfficientNetB0"]
roc_scores = [0.821, 0.737, 0.793, 0.778, 0.795]

plt.figure(figsize=(18, 6))

# Plotting the first subplot (Accuracy)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.bar(models_acc, accuracies, color="lightgreen")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Comparison of Model Accuracies")
plt.ylim(0.8, 1.0)  # Set y-axis limits
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Plotting the second subplot (ROC Scores)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.bar(models_roc, roc_scores, color="lightpink")
plt.xlabel("Models")
plt.ylabel("ROC Score")
plt.title("Comparison of Model ROC Scores")
plt.ylim(0.6, 1.0)  # Set y-axis limits
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()  # Adjust the layout to prevent overlapping
plt.show()
```
